```python

model_name = 'CESM1_pi'
fct = xr.open_mfdataset('/glade/work/jjeffree/results/multiindex/'+model_name+'_tos_zos/*.nc')
mse = ((fct.verification-fct.forecast)**2).mean(('Y','pred_SMILE_M','lib_mi')).load()
r = xr.corr(fct.verification,fct.forecast,('Y','pred_SMILE_M','lib_mi')).load()
fct.close()
print(model_name) 
```

```python
_,phantom_ax = plt.subplots(1,1,figsize=(1,1))
n=7
h=1
fig, axs = plt.subplots(h,n,figsize=(n*4,8.2*h*1.5),sharey=True,sharex=True)


pt.plot.incremental_sailboat(r.sel(index_lon='nino34'),N=91*pt.n_ensemble_members[model_name]*((pt.n_ensemble_members[model_name])//5-1),
                 start_mask=('30P','30P','30P','10P','10P','10P',),
                 later_mask=('30P30A','30P30I','60P','30P','10-30P','30-10P'),
                 fig=fig,axs=axs,cb_axs=axs)
# old_title = axs[i].get_title()
# axs[i].set_title(model_name+'\n'+old_title)
```
