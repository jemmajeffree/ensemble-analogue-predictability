```python
fct = xr.open_mfdataset('/glade/work/jjeffree/results/multiindex/'+model_name+'_tos_zos/*.nc')
```

```python
mse = ((fct.verification-fct.forecast)**2).mean(('Y','pred_SMILE_M','lib_mi')).load()
r = xr.corr(fct.verification,fct.forecast,('Y','pred_SMILE_M','lib_mi')).load()
```

```python
fct
```

```python
fig, ax = plt.subplots(1,3,figsize=(14,10))
for i in range(3):
    plt.sca(ax[i])
    r.isel(index_lon=i).T.plot(vmin=0,vmax=1)
```

```python
fig, ax = plt.subplots(1,3,figsize=(14,10))
for i in range(3):
    plt.sca(ax[i])
    r1 =r.isel(index_lon=i).sel(mask='30P30A')
    r2 =r.isel(index_lon=i).sel(mask='30P')
    np.tanh(np.arctanh(r1)-np.arctanh(r2)).T.plot(vmin=-0.1,vmax=0.1,cmap='BrBG')
```

```python
pt.plot.sailboat(mse,skill_type='mse',N=90*40*5,vlim=(-10,10))
```
